```python
import numpy as np
import cv2


from scipy import ndimage

def enhance_image(image, C=1, K0=0, K1=0.25, K2=0, K3=0.1, neighborhood_size=3):
    # Compute global mean and standard deviation
    mG = np.mean(image)
    σG = np.std(image)

    # Compute local mean and standard deviation using a square neighborhood
    neighborhood = np.ones((neighborhood_size, neighborhood_size))
    msxy = ndimage.convolve(image, neighborhood) / neighborhood_size ** 2
    σsxy = np.sqrt(ndimage.convolve(np.square(image - msxy), neighborhood) / neighborhood_size ** 2)

    # Compute enhanced image using the given formula
    g = np.zeros_like(image)
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            if K0 * mG <= msxy[i, j] <= K1 * mG and K2 * σG <= σsxy[i, j] <= K3 * σG:
                g[i, j] = C * image[i, j]
            else:
                g[i, j] = image[i, j]

    return g


img = cv2.imread(r'.\data\squares.tif', cv2.IMREAD_GRAYSCALE)
enhanced_img = np.zeros_like(img)

K0 = 0
K1 = 0.25
K2 = 0
K3 = 0.1
C = 23.0

Imgfilter=enhance_image(img, C, K0, K1, K2, K3, neighborhood_size=5)



plt.imshow(Imgfilter, cmap='gray')
plt.title("Filter")

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np

# Load the image
image = cv2.imread(r".\data\road.jpg")

# Convert the image to HSV color space
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Define the range of blue color in HSV
lower_blue = np.array([20, 60, 40])
upper_blue = np.array([130, 255, 255])

# Create a mask for the blue board
mask = cv2.inRange(hsv, lower_blue, upper_blue)

# Perform morphological operations to remove noise
kernel = np.ones((3,3),np.uint8)

opening = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(opening, cv2.MORPH_CLOSE, kernel)

# Find contours in the image
contours, hierarchy = cv2.findContours(closing, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Loop over the contours and find the blue board that has text and arrow signs
black_images = []
for contour in contours:
    area = cv2.contourArea(contour)
    if area > 1000:
        x,y,w,h = cv2.boundingRect(contour)
        # Create a black image of the same size as the original image
        black_image = np.zeros_like(image)
        # Draw the blue board on the black image
        cv2.rectangle(black_image,(x,y),(x+w,y+h),(255,255,255),-1)
        # Add the black image to the list
        black_images.append(black_image)

# Combine all the black images using bitwise OR operation
if black_images:
    black_image_combined = cv2.bitwise_or(*black_images)
    # Mask the original image with the black image to make the rest of the pixels black
    result = cv2.bitwise_and(image, black_image_combined)
else:
    result = image



plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))

plt.title("Road Sign board")
plt.tight_layout()
plt.show()
```

# b Write a program that should be able to find the background color of the sign board. It should be a generic function to identify three different primary colors: red, green or blue. It should display blue if the background color of the board is blue.

```python
from PIL import Image

def find_background_color(image_path):
    image = Image.open(image_path)
    pixels = image.load()
    width = image.size[0]
    height = image.size[1]
    blue_count = 0
    for y in range(height):
        for x in range(width):
            r, g, b = pixels[x, y]
            if b > r and b > g:
                blue_count += 1
    if blue_count > (width * height) / 2:
        return "blue"
    else:
        return "not blue"
```

```python
import cv2
import numpy as np
from scipy.signal import find_peaks

def apply_thresholding(image_path):
    img = cv2.imread(image_path, 0)
    threshold_value = find_threshold_value(img)
    ret, thresh = cv2.threshold(img, threshold_value, 255, cv2.THRESH_BINARY)
    thresh = cv2.bitwise_not(thresh)
    return thresh

def find_threshold_value(img):
    hist = cv2.calcHist([img], [0], None, [256], [0, 256])
    peaks, _ = find_peaks(hist.flatten(), distance=1, height=100)
    threshold_value = int(np.mean(peaks))
    return threshold_value



img_path = r".\data\road.jpg"
result = apply_thresholding(img_path)
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title("Result ")
plt.show()
```

```python
def findLines(img, angle):
    # Convert image to grayscale if not already in grayscale
    if len(img.shape) > 2:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Calculate gradient magnitude and direction
    sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
    sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
    mag = cv2.magnitude(sobelx, sobely)
    angle_rad = np.arctan2(sobely, sobelx)
    angle_deg = np.rad2deg(angle_rad)

    # Define threshold for angle
    angle_thresh = 5

    # Define kernels for different orientations
    if angle == 0:
        kernel = np.array([[-1, 2, -1],
                           [-1, 2, -1], 
                           [-1, 2, -1]])
        angle_min = -angle_thresh
        angle_max = angle_thresh
    elif angle == 90:
        kernel = np.array([[-1, -1, -1], 
                           [2, 2, 2], 
                           [-1, -1, -1]])
        angle_min = 90 - angle_thresh
        angle_max = 90 + angle_thresh
    elif angle == 45:
        kernel = np.array([[-1, -1, 2], 
                           [-1, 2, -1], 
                           [2, -1, -1]])
        angle_min = 45 - angle_thresh
        angle_max = 45 + angle_thresh
    elif angle == 135:
        kernel = np.array([[2, -1, -1], 
                           [-1, 2, -1], 
                           [-1, -1, 2]])
        angle_min = 135 - angle_thresh
        angle_max = 135 + angle_thresh

    # Filter image with kernel and angle threshold
    filtered_mag = cv2.filter2D(mag, -1, kernel)
    filtered_angle = np.where((angle_deg >= angle_min) & (angle_deg <= angle_max), mag, 0)

    # Threshold the image to convert it to black and white
    _, thresh = cv2.threshold(filtered_angle, 50, 255, cv2.THRESH_BINARY)


    plt.imshow(thresh, cmap='gray')
    plt.title("Line Filter (Orientation={})".format(angle))
    plt.tight_layout()
    plt.show()


```

# a. Write your own custom function that should be able to do histogram equalization. You should not use any library to perform histogram equalization.

```python
import numpy as np
import matplotlib.pyplot as plt

def histogram_equalization(image):
    
    
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title("Input image")

    plt.tight_layout()
    plt.show()
    
    # Flatten the image into a 1D array
    pixels = image.flatten()

    # Calculate the histogram of the flattened image
    histogram, bins = np.histogram(pixels, bins=256, range=(0, 255))
    

    # Draw the histogram
    
    plt.plot(histogram, color='gray')
    plt.xlabel('Intensity')
    plt.ylabel('Frequency')
    plt.title('Input Image Histogram')
    plt.show()

    # Calculate the cumulative distribution function (CDF) of the histogram
    cdf = histogram.cumsum()

    # Normalize the CDF to have values between 0 and 1
    cdf = (cdf - cdf.min()) * 255 / (cdf.max() - cdf.min())

    # Interpolate the CDF to get the new pixel values
    new_pixels = np.interp(pixels, bins[:-1], cdf)

    # Reshape the 1D array back into the original image shape
    new_image = new_pixels.reshape(image.shape)
    
    
    # Calculate the histogram of the flattened image
    eq_histogram, eq_bins = np.histogram(new_image, bins=256, range=(0, 255))
    
    # Draw the histogram
    
    plt.plot(eq_histogram, color='gray')
    plt.xlabel('Intensity')
    plt.ylabel('Frequency')
    plt.title('histogram equalization.')
    plt.show()
    
    
    
    new_image = np.uint8(new_image)

    # Return the new image
    return new_image
```

```python

def compare_images(img1, img2):
    
    print("Image 1 shape",img1.shape)
    print("Image 2 shape",img2.shape)
    
    img1 = cv2.resize(img1, (256, 256))
    img2 = cv2.resize(img2, (256, 256))

    plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
    plt.title("Input image 1")
    plt.show()
    
    plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
    plt.title("Input image 2")
    
    plt.show()
    # Convert images to grayscale
    img1_gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
    img2_gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)



    # Normalize images
#     img1_norm = cv2.normalize(img1_gray, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
#     img2_norm = cv2.normalize(img2_gray, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)

    # Calculate histograms
    hist1 = cv2.calcHist([img1_gray], [0], None, [256], [0, 256])
    hist2 = cv2.calcHist([img2_gray], [0], None, [256], [0, 256])
    
    
    
    # Plot the histograms using Matplotlib
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
    ax1.hist(img1_gray.ravel(), 256, [0, 256], color='blue')
    ax1.set_xlabel('Pixel intensity')
    ax1.set_ylabel('Frequency')
    ax1.set_title('Image 1 Histogram')
    ax1.set_xlim([0, 256])
    ax2.hist(img2_gray.ravel(), 256, [0, 256], color='red')
    ax2.set_xlabel('Pixel intensity')
    ax2.set_ylabel('Frequency')
    ax2.set_title('Image 2 Histogram')
    ax2.set_xlim([0, 256])
    plt.show()
    
    equ_1 = cv2.equalizeHist(img1_gray)
    equ_2 = cv2.equalizeHist(img2_gray)
    

    # Plot the histograms using Matplotlib
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
    ax1.hist(equ_1.ravel(), 256, [0, 256], color='blue')
    ax1.set_xlabel('Pixel intensity')
    ax1.set_ylabel('Frequency')
    ax1.set_title('Equalized Image 2 Histogram')
    ax1.set_xlim([0, 256])
    
    ax2.hist(equ_2.ravel(), 256, [0, 256], color='red')
    ax2.set_xlabel('Pixel intensity')
    ax2.set_ylabel('Frequency')
    ax2.set_title('Equalized Image 1 Histogram')
    ax2.set_xlim([0, 256])
    plt.show()

    # Calculate similarity score

    similarity = np.sum(np.minimum(hist1, hist2)) / np.sum(hist1)

    

    # Return similarity score and result
    if similarity > 0.8:
        return similarity, "Similar"
    else:
        return similarity, "Different"
```

```python
def calculate_cdf(histogram):

    # Get the cumulative sum of the elements
    cdf = histogram.cumsum()
 
    # Normalize the cdf
    normalized_cdf = cdf / float(cdf.max())
 
    return normalized_cdf
 
def calculate_lookup(src_cdf, ref_cdf):

    lookup_table = np.zeros(256)
    lookup_val = 0
    for src_pixel_val in range(len(src_cdf)):
        lookup_val
        for ref_pixel_val in range(len(ref_cdf)):
            if ref_cdf[ref_pixel_val] >= src_cdf[src_pixel_val]:
                lookup_val = ref_pixel_val
                break
        lookup_table[src_pixel_val] = lookup_val
    return lookup_table
 
def match_histograms(src_image, ref_image):
  
    # Split the images into the different color channels
    # b means blue, g means green and r means red
    src_b, src_g, src_r = cv2.split(src_image)
    ref_b, ref_g, ref_r = cv2.split(ref_image)
 
    # Compute the b, g, and r histograms separately
    # The flatten() Numpy method returns a copy of the array c
    # collapsed into one dimension.
    src_hist_blue, bin_0 = np.histogram(src_b.flatten(), 256, [0,256])
    src_hist_green, bin_1 = np.histogram(src_g.flatten(), 256, [0,256])
    src_hist_red, bin_2 = np.histogram(src_r.flatten(), 256, [0,256])    
    ref_hist_blue, bin_3 = np.histogram(ref_b.flatten(), 256, [0,256])    
    ref_hist_green, bin_4 = np.histogram(ref_g.flatten(), 256, [0,256])
    ref_hist_red, bin_5 = np.histogram(ref_r.flatten(), 256, [0,256])
 
    # Compute the normalized cdf for the source and reference image
    src_cdf_blue = calculate_cdf(src_hist_blue)
    src_cdf_green = calculate_cdf(src_hist_green)
    src_cdf_red = calculate_cdf(src_hist_red)
    ref_cdf_blue = calculate_cdf(ref_hist_blue)
    ref_cdf_green = calculate_cdf(ref_hist_green)
    ref_cdf_red = calculate_cdf(ref_hist_red)
 
    # Make a separate lookup table for each color
    blue_lookup_table = calculate_lookup(src_cdf_blue, ref_cdf_blue)
    green_lookup_table = calculate_lookup(src_cdf_green, ref_cdf_green)
    red_lookup_table = calculate_lookup(src_cdf_red, ref_cdf_red)
 
    # Use the lookup function to transform the colors of the original
    # source image
    blue_after_transform = cv2.LUT(src_b, blue_lookup_table)
    green_after_transform = cv2.LUT(src_g, green_lookup_table)
    red_after_transform = cv2.LUT(src_r, red_lookup_table)
 
    # Put the image back together
    image_after_matching = cv2.merge([
        blue_after_transform, green_after_transform, red_after_transform])
    image_after_matching = cv2.convertScaleAbs(image_after_matching)
 
    return image_after_matching
```

#### Reason for the failure of global thresholding on paper text images, it is because of the variability in the background color and lighting conditions in paper text images. This makes it difficult to choose a single threshold value that can separate the foreground (text) from the background accurately.

```python
# Read the image in grayscale
img = cv2.imread(r".\data\text2.tif", 0)

# Define the size of the window for local thresholding
window_size = 5

# Define a custom function for local thresholding
def local_threshold(img, window_size):
    thresholded_img = np.zeros_like(img)
    pad_size = window_size // 2
    for i in range(pad_size, img.shape[0]-pad_size):
        for j in range(pad_size, img.shape[1]-pad_size):
            window = img[i-pad_size:i+pad_size+1, j-pad_size:j+pad_size+1]
            threshold_value = np.mean(window) - 10
            if img[i, j] > threshold_value:
                thresholded_img[i, j] = 255
    return thresholded_img

# Apply local thresholding using the custom function
Local_thresholded_img = local_threshold(img, window_size)

# Display the thresholded image

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Input Image")
plt.show()

plt.imshow(cv2.cvtColor(Local_thresholded_img, cv2.COLOR_BGR2RGB))
plt.title("local thresholding ")
plt.show()
```
